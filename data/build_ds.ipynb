```python
from datasets import load_dataset, load_from_disk
import matplotlib.pyplot as plt
import random
```

```python
def show_samples(dataset, n=5):
    """Display n random samples from the dataset."""
    indices = random.sample(range(len(dataset)), n)
    for i in indices:
        print("Image:", dataset[i]['image'])
        plt.imshow(dataset[i]['image'])
        plt.axis('off')
        plt.show()
```

### African atire dataset
The African Attire Images Dataset is a curated collection of images showcasing various African traditional attires and clothing styles.

```python
show_samples(african_atire['train'], n=5)
```
